```python
# function with no argument and no return type

l = int(input("Enter l = "))                #global variable
b = int(input("Enter b = "))                #global variable

def cal(): 
    a = l*b
    print(a)
    
cal()
print(l,b)
```

```python
# function with no argument and no return type

def cal():
    global l,b,a
    l = int(input("Enter l = "))                #local variable
    b = int(input("Enter b = "))                #local variable
    a = l*b
    print(a)
    
cal()
print(l,b)
```

```python
# function with argument and no return type
def cal(l,b):              #parameter
    a = l*b
    print(a)

l = int(input("Enter l = "))
b = int(input("Enter b = "))
cal(l,b)                   #argument
```

```python
# function with argument and no return type
def cal(x,y):              #parameter
    a = x*y
    print(a)

l = int(input("Enter l = "))
b = int(input("Enter b = "))
cal(l,b)                   #argument
```

```python
# function with argument and no return type
def cal(data):              #parameter
    x,y,z = data
    v = x*y*z
    print(v)

l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))

data = (l,b,h)

cal(data)                   #argument
```

```python
# function with no argument and return type
def cal():
    l = 10
    b = 5
    a = l*b
    return a

print(cal())
```

```python
# function with no argument and return type
def cal():
    l = 10
    b = 5
    a = l*b
    return a

h = 2
area = cal()
volume = area*h
print(volume)
```

```python
# function with no argument and return type
def cal():
    l = 10
    b = 5
    h = 2
    a = l*b
    v = a*h
    return a,v

x = cal()
ar,vo = x
print("The area = ",ar)
print("The volume = ",vo)
```

```python
# function with argument and return type
def cal(l,b,h):
    a = l*b
    v = a*h
    return a,v

l = 10
b = 5
h = 2

x = cal(l,b,h)
ar,vo = x
print("The area = ",ar)
print("The volume = ",vo)
```

```python
def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + - * /  ")
    if o == '+':
        print(a+b)
    elif o == '-':
        print(a-b)
    elif o == '*':
        print(a*b)
    elif o == '/':
        print(a/b)
    else:
        print("Invalid operator")
    
    x = input("Enter y for more cal ")
    if x == 'y':
        cal()
        
cal()
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + - * /  ")
    if o == '+':
        print(add(a,b))
    elif o == '-':
        print(sub(a,b))
    elif o == '*':
        print(mul(a,b))
    elif o == '/':
        print(div(a,b))
    else:
        print("Invalid operator")
    
    x = input("Enter y for more cal ")
    if x == 'y':
        cal()
        
cal()
```

```python
bill = ""
n = int(input("Enter n = "))
def billing():
    global bill
    product = input("Enter product name = ")
    price = int(input("Enter price = "))
    quantity = int(input("Enter quantity = "))
    total = price*quantity
    info = f"{product} {price} {quantity} {total}\n"
    bill = bill +info
for i in range(n):   
    billing()
    
print(bill)
```

```python
bill = ""
grand_total = 0
n = int(input("Enter n = "))
def billing(product,price,quantity,total):
    global bill,grand_total
    info = f"{product} {price} {quantity} {total}\n"
    bill = bill +info
    grand_total = grand_total+total
    
    
for i in range(n):
    product = input("Enter product name = ")
    price = int(input("Enter price = "))
    quantity = int(input("Enter quantity = "))
    total = price*quantity
    billing(product,price,quantity,total)
    
print(bill)
print(grand_total)
```
